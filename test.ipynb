## Data example

```python
import pandas as pd
import sqlite3
from pathlib import Path
```

```python
conn = sqlite3.connect(":memory:")

data_dir = Path("data")

csv_files = [
    "cart_data.csv",
    "cart_item_data.csv",
    "category_data.csv",
    "data.csv",
    "order_data.csv",
    "order_items_data.csv",
    "payment_data.csv",
    "product_data.csv",
    "review_data.csv",
    "shipping_data.csv",
    "user_data.csv",
]

for file in csv_files:
    table_name = file.replace(".csv", "")
    df = pd.read_csv(data_dir / file)
    df.to_sql(table_name, conn, index=False, if_exists="replace")
    print(f"Loaded {file} as table: {table_name}")
```

## Primary keys
Categories.category_id

Products.product_id

Users.user_id

Orders.order_id

Order_Items.order_item_id

Reviews.review_id

Cart.cart_id

Cart_Items.cart_item_id

Payments.payment_id

Shipping.shipping_id

## Foreign keys and join paths
Products.category_id       → Categories.category_id

Orders.user_id             → Users.user_id

Order_Items.order_id       → Orders.order_id

Order_Items.product_id     → Products.product_id

Reviews.user_id            → Users.user_id

Reviews.product_id         → Products.product_id

Cart.user_id               → Users.user_id

Cart_Items.cart_id         → Cart.cart_id

Cart_Items.product_id      → Products.product_id

Payments.order_id          → Orders.order_id

Shipping.order_id          → Orders.order_id
